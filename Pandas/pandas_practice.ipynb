```python

import seaborn as sns
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(14, 10))
axes = axes.flat
columnas_numeric = Store.select_dtypes(include=['float64', 'int64']).columns

for i, colum in enumerate(columnas_numeric):
    sns.histplot(
        data    = df,
        x       = colum,
        stat    = "count",
        kde     = True,
        color   = (list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
        line_kws= {'linewidth': 2},
        alpha   = 0.3,
        ax      = axes[i]
    )
    axes[i].set_title(colum, fontsize = 9, fontweight = "bold")
    axes[i].tick_params(labelsize = 8)
    axes[i].set_xlabel("")
    
    
fig.tight_layout()
plt.subplots_adjust(top = 0.9)
fig.suptitle('Distribución variables numéricas', fontsize = 14, fontweight = "bold");
```

```python
#instead of using all columns in  datframe usly only categorical once and numerical once to check outliners



import matplotlib.pyplot as plt
import seaborn as sns

columnas = df.columns  # Assuming you want to plot columns from 'df'
n = 1
plt.figure(figsize=(20, 15))

for column in columnas:
    if n > 16:  # Stop when you reach the last subplot (16)
        break
    plt.subplot(4, 4, n)
    n += 1  # Increment by 1 to move to the next subplot
    sns.boxplot(df[column])
    plt.tight_layout()  # Adjust the layout to avoid overlapping plots

plt.show()
```
