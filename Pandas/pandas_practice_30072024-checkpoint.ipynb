```python
import pandas as pd

# Sample DataFrame for demonstration
table = pd.DataFrame({
    'Order Date': ['2016-11-07', '2016-11-08', '2015-10-18'],
    'Ship Date': ['2015-10-17', '2015-10-18', '2015-10-19'],
    'Ship Mode': ['Standard Class', 'Second Class', 'Standard Class']
})

# Convert to datetime
table['Order Date'] = pd.to_datetime(table['Order Date'])
table['Ship Date'] = pd.to_datetime(table['Ship Date'])

# Filter DataFrame
result = table[(table['Order Date'] >= '2016-11-08') & 
               (table['Ship Date'] <= '2015-10-18') & 
               (table['Ship Mode'].isin(['Standard Class', 'Second Class']))]['Order Date']

print(result)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Get all numeric columns from the DataFrame
numeric_columns = table.select_dtypes(include="number").columns

# Determine the number of rows and columns needed for the subplots
num_plots = len(numeric_columns)
num_cols = 2  # Number of columns for subplots
num_rows = math.ceil(num_plots / num_cols)  # Calculate the required number of rows

plt.figure(figsize=(15, 5 * num_rows))  # Adjust the figure size based on the number of rows

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.boxplot(x=table[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()  # Adjust layout so everything fits without overlapping
plt.show()  # Display the plots
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Get all categorical (object) columns from the DataFrame
numeric_columns = table.select_dtypes(include="object").columns

# Determine the number of rows and columns needed for the subplots
num_plots = len(numeric_columns)
num_cols = 2  # Number of columns for subplots
num_rows = math.ceil(num_plots / num_cols)  # Calculate the required number of rows

plt.figure(figsize=(18, 6 * num_rows))  # Increase the figure size for better readability

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.countplot(x=table[col])  # Use countplot for categorical data
    plt.title(f'Countplot of {col}', fontsize=16)  # Increase title font size
    plt.xlabel(col, fontsize=14)  # Increase x-axis label font size
    plt.ylabel('Count', fontsize=14)  # Increase y-axis label font size
    plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-ticks and adjust font size for readability

plt.tight_layout()  # Adjust layout so everything fits without overlapping
plt.show()  # Display the plots

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Get all categorical (object) columns from the DataFrame
categorical_columns = table.select_dtypes(include="object").columns

# Determine the number of rows and columns needed for the subplots
num_plots = len(categorical_columns)
num_cols = 2  # Number of columns for subplots
num_rows = math.ceil(num_plots / num_cols)  # Calculate the required number of rows

plt.figure(figsize=(20, 8 * num_rows))  # Increase the figure size further

for i, col in enumerate(categorical_columns, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.countplot(x=table[col])  # Use countplot for categorical data
    plt.title(f'Countplot of {col}', fontsize=20)  # Increase title font size
    plt.xlabel(col, fontsize=16)  # Increase x-axis label font size
    plt.ylabel('Count', fontsize=16)  # Increase y-axis label font size
    plt.xticks(rotation=45, ha='right', fontsize=14)  # Rotate x-ticks and adjust font size

plt.subplots_adjust(hspace=0.4, wspace=0.4)  # Adjust spacing between plots
plt.show()  # Display the plots
```
