```python
# Function to check if a value is not numeric and not a specific unwanted string
def is_not_numeric(value):
    # Define unwanted strings
    unwanted_strings = {"False", "NaN", "True"}
    
    # Check if value is not numeric and not an unwanted string
    if isinstance(value, (int, float, np.number)):
        return False
    elif isinstance(value, str) and value in unwanted_strings:
        return False
    else:
        return True

# Applying the function to the DataFrame
non_numeric_mask = Numerical_1.applymap(is_not_numeric)

# Printing the mask of non-numeric values
print("Mask of Non-Numeric Values:")
print(non_numeric_mask)

# Optionally, filter out non-numeric values
non_numeric_df = Numerical_1.where(non_numeric_mask, other=np.nan)
print("\nNon-Numeric Values in DataFrame:")
print(non_numeric_df)
```

```python
#to get all corelated values but something is pending like sorting order
import pandas as pd

# Assuming `combined` is your DataFrame
# Step 1: Calculate the correlation matrix
Numerical_2_corr = combined.corr()

# Step 2: Flatten the correlation matrix and sort by absolute values in descending order
corr_pairs = Numerical_2_corr.unstack().reset_index()
corr_pairs.columns = ['Feature_1', 'Feature_2', 'Correlation']

# Filter out self-correlations (correlation of a feature with itself)
corr_pairs = corr_pairs[corr_pairs['Feature_1'] != corr_pairs['Feature_2']]

# Step 3: Sort the feature names in each row so that (Feature_1, Feature_2) and (Feature_2, Feature_1) are treated as the same pair
corr_pairs['sorted_features'] = corr_pairs.apply(lambda x: tuple(sorted([x['Feature_1'], x['Feature_2']])), axis=1)

# Step 4: Drop duplicates based on the sorted feature pairs
unique_corr_pairs = corr_pairs.drop_duplicates(subset='sorted_features').drop(columns='sorted_features')

# Step 5: Sort by absolute correlation values in descending order
sorted_corr_pairs = unique_corr_pairs.sort_values(by='Correlation', ascending=False, key=abs)

# Display the sorted and unique correlation values
print(sorted_corr_pairs.head(200))

# Step 3: Get unique columns based on sorted correlations
sorted_columns = pd.unique(sorted_corr_pairs[['Feature_1', 'Feature_2']].values.ravel('K'))
print("Columns in order of descending correlations:")
print(sorted_columns)

```
