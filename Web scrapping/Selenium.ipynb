```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

# Setup ChromeDriver
chrome_driver_path = r"C:\Users\chaitanya\.cache\selenium\chromedriver\win64\128.0.6613.86\chromedriver.exe"
service = Service(chrome_driver_path)
options = Options()
options.add_argument("--headless")  # Run in headless mode
driver = webdriver.Chrome(service=service, options=options)

# Open the webpage
driver.get("https://www.igp.com/gifts-for-friend")

# Wait for the page to fully load initially
time.sleep(5)

# Scroll and load all items
last_height = driver.execute_script("return document.body.scrollHeight")
new_items_loaded = True
scroll_attempts = 0
product_set = set()  # To keep track of unique products

while new_items_loaded:
    # Scroll down to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)  # Wait for new products to load
    
    # Capture current products
    gift_containers = driver.find_elements(By.CLASS_NAME, 'product-item')
    
    for container in gift_containers:
        # Extract gift name
        try:
            name_tag = container.find_element(By.CSS_SELECTOR, 'div.product-name-w p.product-name')
            name = name_tag.text.strip()
        except NoSuchElementException:
            name = 'No Name'
        
        # Extract gift price
        price = 'No Price'  # Default value
        try:
            price_tag = container.find_element(By.CSS_SELECTOR, 'p.product-price span')
            price = price_tag.text.strip()
        except NoSuchElementException:
            try:
                price_tag = container.find_element(By.CSS_SELECTOR, 'div.price span')
                price = price_tag.text.strip()
            except NoSuchElementException:
                try:
                    price_tag = container.find_element(By.XPATH, ".//*[contains(text(),'₹')]")
                    price = price_tag.text.strip()
                except NoSuchElementException:
                    price = 'No Price'

        product_info = (name, price)
        
        # Add to set if new
        if product_info not in product_set:
            product_set.add(product_info)
            print(f"Product Name: {name}, Product Price: {price}")
    
    # Get new height after scrolling
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    # Check if the scroll height has changed
    if new_height == last_height:
        scroll_attempts += 1
        if scroll_attempts > 3:  # If scrolled 3 times without new items, stop
            new_items_loaded = False
    else:
        last_height = new_height
        scroll_attempts = 0  # Reset scroll attempts

# Close the browser
driver.quit()
```
